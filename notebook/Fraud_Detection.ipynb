```python
X = df.drop('Class', axis=1)

y = df['Class']
```

```python
print(X.shape)

print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
print(y_train.value_counts())

print(y_test.value_counts())
```

```python
pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_train
)
```

```python
print(y_train.value_counts())

print(y_train_smote.value_counts())
```

```python
print(df.shape)
print(df['Class'].value_counts())
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42)

lr.fit(X_train_smote, y_train_smote)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred_lr))

print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Precision:", precision_score(y_test, y_pred_lr))
print("Recall:", recall_score(y_test, y_pred_lr))
print("F1 Score:", f1_score(y_test, y_pred_lr))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train_smote, y_train_smote)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall:", recall_score(y_test, y_pred_rf))
print("F1 Score:", f1_score(y_test, y_pred_rf))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
import shap
import numpy as np
import matplotlib.pyplot as plt
```

```python
explainer = shap.TreeExplainer(rf)
```

```python
X_test_sample = X_test.sample(1000, random_state=42)
```

```python
shap_values = explainer.shap_values(X_test_sample)
```

```python
print(type(shap_values))
print(np.array(shap_values).shape)
```

```python
shap.summary_plot(
    shap_values[:, :, 1],
    X_test_sample
)
```

```python
fraud_indices = y_test[y_test == 1].index

print("Number of fraud cases:", len(fraud_indices))
```

```python
fraud_sample = X_test.loc[[fraud_indices[0]]]

fraud_sample
```

```python
fraud_shap = explainer.shap_values(fraud_sample)
```

```python
shap.plots.waterfall(
    shap.Explanation(
        values=fraud_shap[0][:, 1],
        base_values=explainer.expected_value[1],
        data=fraud_sample.iloc[0],
        feature_names=fraud_sample.columns
    )
)
```
